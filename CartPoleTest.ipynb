```python
import numpy as np
import gymnasium as gym
import joblib
import os
import matplotlib.pyplot as plt
from gym_wrapper import GymWrapper
import config

def run_stress_test(config, test_type='mass'):
    # Load your trained model
    model_path = os.path.join(config.LOG_PARAMS['save_path'], 'trained_dqn_model_with_params.pkl')
    model_info = joblib.load(model_path)
    controller = model_info['model']

    if hasattr(controller.model, 'eval'):
        controller.model.eval()

    results = []

    # Define the range for the stress test
    # We will test pole mass from 0.1 (standard) up to 2.0 (very heavy)
    test_range = np.linspace(0.1, 2.0, 10) if test_type == 'mass' else np.linspace(10.0, 2.0, 10)

    print(f"Starting Stress Test on: {test_type}")

    for val in test_range:
        # Create environment with MODIFIED physics
        if test_type == 'mass':
            # Create standard env then manually override the mass
            raw_env = gym.make('CartPole-v1')
            raw_env.unwrapped.masspole = val
            raw_env.unwrapped.total_mass = raw_env.unwrapped.masscart + val
            raw_env.unwrapped.polemass_length = val * raw_env.unwrapped.length
        else:
            # Testing Reduced Force (Motor Failure simulation)
            raw_env = gym.make('CartPole-v1')
            raw_env.unwrapped.force_mag = val

        env = GymWrapper(raw_env, mode='4D', action_mode='discrete')

        episode_rewards = []
        for _ in range(10): # Run 10 trials per setting for an average
            state, _ = env.reset()
            done = False
            total_r = 0
            while not done:
                action = controller.get_action(state, explore=False)
                state, reward, terminated, truncated, _ = env.step(action)
                done = terminated or truncated
                total_r += reward
            episode_rewards.append(total_r)

        avg_reward = np.mean(episode_rewards)
        results.append(avg_reward)
        print(f"{test_type} value: {val:.2f} | Avg Reward: {avg_reward}")
        env.close()

    return test_range, results

# Execute the test
mass_range, mass_scores = run_stress_test(config, test_type='mass')

# Plotting the "Breaking Point"
plt.figure(figsize=(10, 5))
plt.plot(mass_range, mass_scores, marker='o', color='red')
plt.title('DQN Robustness: Impact of Increasing Pole Mass')
plt.xlabel('Pole Mass (kg) [Standard = 0.1]')
plt.ylabel('Average Reward (Max 500)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import gymnasium as gym
import joblib
import os
import matplotlib.pyplot as plt
from gym_wrapper import GymWrapper
from pyvirtualdisplay import Display
from gymnasium.wrappers import RecordVideo
import config

def stress_test_initial_angle(config, num_trials=5):
    # Load Model
    model_path = os.path.join(config.LOG_PARAMS['save_path'], 'trained_dqn_model_with_params.pkl')
    model_info = joblib.load(model_path)
    controller = model_info['model']
    if hasattr(controller.model, 'eval'): controller.model.eval()

    # Define angles from 0.05 (normal) to 0.4 radians (approx 23 degrees - very steep!)
    # Note: CartPole-v1 failure threshold is 0.209 radians (12 degrees),
    # but we want to see if it can recover even when starting PAST the fail point.
    angle_thresholds = np.linspace(0.01, 0.20, 10)
    avg_rewards = []

    print("Starting Initial Angle Stress Test...")

    for angle in angle_thresholds:
        raw_env = gym.make('CartPole-v1')
        env = GymWrapper(raw_env, mode='4D', action_mode='discrete')

        trial_rewards = []
        for _ in range(num_trials):
            # Manually reset with a specific extreme angle
            state, _ = env.reset()
            # Overwrite the pole angle (index 2 in 4D state) with the test angle
            # We alternate left and right tilt
            side = 1 if np.random.random() > 0.5 else -1
            env.env.unwrapped.state[2] = angle * side

            done = False
            total_r = 0
            while not done:
                action = controller.get_action(env._filter_state(env.env.unwrapped.state), explore=False)
                state, reward, terminated, truncated, _ = env.step(action)
                done = terminated or truncated
                total_r += reward
            trial_rewards.append(total_r)

        avg_rewards.append(np.mean(trial_rewards))
        print(f"Start Angle: {np.degrees(angle):.2f}Â° | Avg Reward: {np.mean(trial_rewards)}")
        env.close()

    return angle_thresholds, avg_rewards

# Execute and Plot
angles, scores = stress_test_initial_angle(config)

plt.figure(figsize=(10, 5))
plt.plot(np.degrees(angles), scores, marker='s', linestyle='--', color='blue')
plt.title('DQN Robustness: Survival by Starting Pole Angle')
plt.xlabel('Initial Pole Angle (Degrees)')
plt.ylabel('Average Reward')
plt.axvline(x=12, color='red', linestyle=':', label='Default Failure Limit (12Â°)')
plt.legend()
plt.grid(True)
plt.show()
```

# Testing Sarsa

This section is to be used only for testing the Sarsa model and to display and save its results

```python
import matplotlib.pyplot as plt
import numpy as np
import joblib
import os

def plot_test_results(logger, save_path=None):
    plt.figure(figsize=(12, 6))

    # Plot test rewards
    plt.plot(logger.episode_rewards, label='Test Reward', color='green')

    # Optionally, plot average reward (moving average)
    window_size = 10
    avg_rewards = np.convolve(logger.episode_rewards, np.ones(window_size) / window_size, mode='valid')
    plt.plot(range(len(avg_rewards)), avg_rewards, label='Average Test Reward', color='orange')

    plt.xlabel('Episode')
    plt.ylabel('Reward')
    plt.title('SARSA Test Rewards')
    plt.legend()

    if save_path:
        plt.savefig(save_path)
    plt.show()

# Modify the test_sarsa function to include plotting
def test_sarsa(model_class, config, num_episodes=100):
    env = GymWrapper(gym.make('CartPole-v1'))

    # Load the trained SARSA model
    model_path = os.path.join(config.LOG_PARAMS['save_path'], 'sarsa_model.pkl')
    controller = joblib.load(model_path)

    test_rewards = []
    test_lengths = []

    for episode in range(num_episodes):
        state = env.reset()
        episode_reward = 0
        done = False
        t = 0

        action = controller.get_action(state)  # Initial action

        while not done:
            next_state, reward, done, _ = env.step(action)
            next_action = controller.get_action(next_state)  # Next action

            # No update to Q-table here, just testing
            state = next_state
            action = next_action
            episode_reward += reward
            t += 1

        test_rewards.append(episode_reward)
        test_lengths.append(t)

        print(f"Episode {episode + 1}/{num_episodes}: Reward = {episode_reward}, Length = {t}")

    # Calculate and print the average reward and length
    average_test_reward = sum(test_rewards) / num_episodes
    average_test_length = sum(test_lengths) / num_episodes

    print(f"Average Test Reward over {num_episodes} episodes: {average_test_reward}")
    print(f"Average Test Episode Length: {average_test_length}")

    # Save the test results in logs
    test_logger = DataLogger(config.LOG_PARAMS)
    test_logger.episode_rewards = test_rewards
    test_logger.episode_lengths = test_lengths
    test_logger.save_logs_as_csv(state='test')
    test_logger.save_metrics(state='test')

    # Plot and save the test results
    plot_path = os.path.join(config.LOG_PARAMS['save_path'], 'sarsa_test_plot.png')
    plot_test_results(test_logger, save_path=plot_path)

    env.close()
```
